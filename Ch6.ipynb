#### Use Series obj to build DataFrame obj

```python
import pandas as pd

s1 = pd.Series(["Bike","Bus","Car","Truck"])
s2 = pd.Series([3,4,6,2])
s3 = pd.Series([2,4,4,6])


# s1 = ["Bike","Bus","Car","Truck"]
# s2 = [3,4,6,2]
# s3 = [2,4,4,6]


data = {'': s1, '': s2, '': s3 } 
df = pd.DataFrame(data) 
print(df)
```

#### Use Python dictionary to build DataFrame obj
* Because Series also need to be built by python dictionary to build DataFrame, so we can use python dictionary to build DataFrame obj:

```python
import pandas as pd

data = {'': ["Bike","Bus","Car","Truck"],
        '': [3,4,6,2],
        '': [2,4,4,6] } 
df = pd.DataFrame(data) 
print(df)
```

#### Rearrange the row index & column index 

```python
import pandas as pd

data = {'': ["Bike","Bus","Car","Truck"],
        '': [3,4,6,2],
        '': [2,4,4,6] } 
df = pd.DataFrame(data) 
labels = ["A","B","E","D"]
df.columns = ["Types", "Count", "Wheels"]
labels[2] = "C"
df.index = labels
print(df)
```

#### Use specific column as row index
* DataFrame can use set_index() method to assign row index, reset_index() method can reset to default number as index

```python
import pandas as pd

data = {'': ["Bike","Bus","Car","Truck"],
        '': [3,4,6,2],
        '': [2,4,4,6] } 
df = pd.DataFrame(data) 
df.set_index("", inplace=True)
print(df)
df.reset_index(inplace=True)
print(df)
```

#### Add record & column
* Add record in DataFrame, we can use append() method

```python
import pandas as pd

data = {'': ["Bike","Bus","Car","Truck"],
        '': [3,4,6,2],
        '': [2,4,4,6] } 
df = pd.DataFrame(data) 
s = pd.Series({'':"Bicycle",'':5,'':2}) 
df2 = df.append(s, ignore_index=True)
print(df2.tail())

print("**************")

df[""] = [1, 20, 4, 2]
print(df)
```

#### Remove record & column

```python
import pandas as pd

data = {'': ["Bike","Bus","Car","Truck"],
        '': [3,4,6,2],
        '': [2,4,4,6] } 
df = pd.DataFrame(data) 
df.set_index("", inplace=True)
print(df)
print("**************")
df2 = df.drop(["Bus", "Truck"])
print(df2)
print("**************")
df3 = df.drop(df.index[[0, 2]])
print(df3)
print("**************")
df4 = df.drop([""], axis=1)    # Steven: Probably because there has chance multiple [""] column existed 
print(df4)
```
