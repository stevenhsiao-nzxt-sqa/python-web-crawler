#### Use Series obj to build DataFrame obj

#### Rearrange the row index & column index 

```python
import pandas as pd

data = {'': ["Bike","Bus","Car","Truck"],
        '': [3,4,6,2],
        '': [2,4,4,6] } 
df = pd.DataFrame(data) 
labels = ["A","B","E","D"]
df.columns = ["Types", "Count", "Wheels"]
labels[2] = "C"
# labels[3] = "C"  # 
df.index = labels
print(df)
# print(df.index[0:3])
```

#### Use specific column as row index
* DataFrame can use set_index() method to assign row index, reset_index() method can reset to default number as index

```python
import pandas as pd

data = {'': ["Bike","Bus","Car","Truck"],
        '': [3,4,6,2],
        '': [2,4,4,6] } 
df = pd.DataFrame(data) 
df.set_index("", inplace=True)
print(df)
# print(df.index[0:3])
df.reset_index(inplace=True)
print(df)
```

### 6-1-4 Select, filter & sort DataFrame data

```python
import pandas as pd

data = {'': ["Bike","Bus","Car","Truck"],
        '': [3,4,6,2],
        '': ["2","4","4","6"] } 
df = pd.DataFrame(data, index=["A","B","C","D"]) 
print(df)
```

#### Use column index select one or multiple column

```python
print(df[""])
print("***")
print(df[["", ""]].head(3))
```

#### Use loc & iloc indexer to select data
* df.loc[, ]
* df.iloc[, ]

```python
import pandas as pd

data = {'': ["Bike","Bus","Car","Truck"],
        '': [3,4,6,2],
        '': ["2","4","4","6"] } 
df = pd.DataFrame(data, index=["A","B","C","D"]) 
print(df)
print(df.loc["A", ""])
print("***")
print(df.loc[["C","D"], ["",""]])
print("***")
print(df.loc[:, ["",""]])
print("***")
print(df.loc["B":"C", "":""])
print("***")
print(df.iloc[3])
print("***")
print(df.iloc[2:4, 1:3])
```

#### Use boolean to filter data

```python
import pandas as pd

data = {'': ["Bike","Bus","Car","Truck"],
        '': [3,4,6,2],
        '': ["2","4","4","6"] } 
df = pd.DataFrame(data, index=["A","B","C","D"]) 
df[""] = df[""].astype("int64")
print(df[df. > 3])
```

#### Specify the column index to sort data

```python
import pandas as pd

data = {'': ["Bike","Bus","Car","Truck"],
        '': [3,4,6,2],
        '': ["2","4","4","6"] } 
df = pd.DataFrame(data, index=["A","B","C","D"]) 
df2 = df.sort_values("", ascending=False)
print(df2)

```

#### Data Frame Group Data Usage

```python
# BatchNo,Price,Rev
# a,100,32434
# a,80,16543
# b,120,1564
# b,130,16543
# b,200,5000
# c,360,32434
# c,250,3456


import pandas as pd

df = pd.read_csv("./MP22027_Example/ch06/batchSales.csv")
print(df)
print("=======")

df2 = df.groupby("BatchNo").mean()
print(df2)
print("=======")

df3 = df.groupby("BatchNo").count()
print(df3)
print("=======")

df4 = df.mean() # all avg
print(df4)
```

### 6-4-1 Missing Data Handling
The missing data(NaN) can not be calculate, so need to be handled:
1. Remove()
2. Replenish()

#### Show NaN Info

```python
import pandas as pd

df = pd.read_csv("./MP22027_Example/ch06/titanic_test.csv")
print(df.head())
print("=======")
print(df.info())
```

#### Remove NaN data

```python
import pandas as pd

df = pd.read_csv("./MP22027_Example/ch06/titanic_test.csv")

print(df["Age"].isnull().sum())
df2 = df.dropna(subset=["Age"])
print(len(df2))
```

#### Fill NaN data

```python
import pandas as pd

df = pd.read_csv("./MP22027_Example/ch06/titanic_test.csv")

df["Age"] = df["Age"].fillna(value=100)
print(df["Age"].isnull().sum())
```

### 6-4-2 Duplicate Data Handling

```python
import pandas as pd

df = pd.read_csv("./MP22027_Example/ch06/sales.csv")
print(df)
print("==========")

df1 = df.drop_duplicates("Country")
print(df1)
```

### 6-4-3 Convert Data Category to Value Data

```python
import pandas as pd

df = pd.read_csv("./MP22027_Example/ch06/shoes.csv")
print(df)
print("==========")

size_mapping = {
    "XXL": 5,
    "XL": 4,
    "L": 3,
    "M": 2,
    "S": 1,
    "XS": 0
}

df["Size"] = df["Size"].map(size_mapping)
print(df)
```
