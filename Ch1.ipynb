### 1-3 Function, Module & Package
#### Function & Function call

with/without argument, with 'return' if need return value

```python
def print_msg():
    print("welcome to learn python!")

def is_valid_num(no):
    if no >= 0 and no <= 200.0:
        return True
    else:
        return False

def convert_to_f(c):
    f = (9.0 * c) / 5.0 + 32.0
    return f

print_msg()
c = 100
f = convert_to_f(c)
print("Fahrenheit : " + str(f))
if is_valid_num(c):
    print("legal!")
else:
    print("illegal")
```

## 1-4 Container Type
### 1-4-1 Lists
* similar to other languages' Arrays, and it's Mutable(允許更改)
#### Basic usage:

```python
ls = [6, 4, 5]   
print(ls, ls[2])  
print(ls[-1])    
ls[2] = "py"      
print(ls)         
ls.append("bar")
print(ls)     
ele = ls.pop()
print(ele, ls)
```

#### Slice List:

```python
nums = list(range(5))  
print(nums)            
```

```python
print(nums[2:4])       
print(nums[2:])        
print(nums[:2])        
print(nums[:])         
print(nums[:-1])       
nums[2:4] = [7, 8]     
print(nums)            
```

#### Visit Lists:

```python
animals = ['cat', 'dog', 'bat'] 
for animal in animals:
    print(animal)
```

* If need to show or get the index, can use enumerator()

```python
animals = ['cat', 'dog', 'bat'] 
for index, animal in enumerate(animals):
    print(index, animal)
```

#### List Comprehension
* Use simple & clean syntax to create List:  (Use [] & for loop)

```python
list1 = [x for x in range(10)] 
print("[x for x in range(10)]")
print(str(list1))
```

```python
list2 = [x+1 for x in range(10)]
print("[x+1 for x in range(10)]")
print(str(list2))
list3 = [x for x in range(10) if x % 2 == 0]
print("[x for x in range(10) if x%2==0]")
print(str(list3))
list4 = [x*2 for x in range(10) if x % 2 == 0]
print("[x*2 for x in range(10) if x%2==0]")
print(str(list4))
```

### 1-4-2 Dictionary
* the container type for store key-value data 
#### Basic Use:

```python
d = {"cat": "white", "dog": "black"}  
print(d["cat"])       
print("cat" in d)     
d["pig"] = "pink"     
print(d["pig"])       
print(d.get("monkey", "N/A")) 
print(d.get("pig", "N/A"))     
del d["pig"]          
print(d.get("pig", "N/A"))     

```

#### Visit dictionary

```python
d = {"chicken": 2, "dog": 4, "cat": 4, "spider": 8} 
for animal in d:
    legs = d[animal]
    print(animal, legs)
```

* if need to visit key & value of dictionary, use items()

```python
d = {"chicken": 2, "dog": 4, "cat": 4, "spider": 8} 
for animal, legs in d.items():
    print("animal: %s has %d legs" % (animal, legs))

```

#### dictionary Comprehension

```python
d1 = {x:x*x for x in range(10)} 
print("{x:x*x for x in range(10)}")
print(str(d1))
d2 = {x:x*x for x in range(10) if x % 2 == 1}
print("{x:x*x for x in range(11) if x%2==1}")
print(str(d2))
```

#### Tuple
* Tuple is similar to List, but it's immutable!!!!
* 

```python
t = (5, 6, 7, 8)   
print(type(t))    
print(t)          
print(t[0])       
print(t[1])       
print(t[-1])      
print(t[-2])      
for ele in t:     
    print(ele, end=" ")  
```

```python
# tuple is immutable
t = (5,)
print(t[0])
t[0] = 6
print(t)
```

```python
# list is mutable
t = [5]
print(t[0])
print(id(t))
t[0] = 6
print(t)
print(id(t))
```
