### 2-6 CSV & JSON

#### CSV
CSV(Comma-Separated Values) is a type that use pure text to represent form data


in CSV, will become:

程式語言,開發者,上市年,副檔名

Python,Cuido van Rossum,1991,.py

C++,Bjarne Stroustrup,.cpp

#### Read CSV file

```python
import csv

csvfile = ".\MP22027_Example\ch02\pl.csv"
with open(csvfile, 'r') as fp:
    reader = csv.reader(fp)
    for row in reader:
        print(','.join(row))
```

#### Write CSV file

```python
import csv

csvfile= ".\MP22027_Example\ch02\pl2.csv"
lst1= [["cow", "Gui Luan Mei", "Moo"],
       ["dog","jeff","Wroof"],
       ["cat","Shelly","Meow"]
      ]
with open(csvfile, 'w+', newline='') as fp:
    writer = csv.writer(fp)
    writer.writerow(["Animal","Owner","Bark"])
    for row in lst1:
        writer.writerow(row)
```

#### 2-6-1 JSON
JSON (JavaScript Object Notation), a data exchange format, which use text only to represent JavaScript object.
But since JSON is self-described easy to read and understand, we now use JSON as a not-language-related data exchange format.

#### what is JSON
JSON use curly brakets to define couples of key-value pairs, it's similar to dictionary in Python 

```python
{
    "key1": "value1",
    "key2": "value2",
    "key3": "value3"
}
```

If it's a array of JSON object, we can use "[ ]" as a normal array

```python
[
    {
        "key1": "value1",
        "key2": "value2",
        "key3": "value3"
    },
    {
        "key1": "value1",
        "key2": "value2",
        "key3": "value3"
    }
]
```

#### JSON, Pyton dictionary convert

```python
import json

data = {
    "name": "jeff",
    "grade": "87",
    "tel": "0912345678"
}

json_str = json.dumps(data)
print(json_str)
data2 = json.loads(json_str)
print(data2)
```

The code above,
we use dumps() to convert python dictionary to a json string,
and use loads() to convert a json string back to python dictionary

#### Write JSON data into file

```python
import json

data = {
    "name": "jeff",
    "grade": "87",
    "tel": "0912345678"
}
jsonfile = ".\MP22027_Example\ch02\Student.json"
with open(jsonfile, 'w') as fp:
    json.dump(data, fp)
```

#### Load JSON file

```python
import json

jsonfile = ".\MP22027_Example\ch02\Student.json"
with open(jsonfile, 'r') as fp:
    data = json.load(fp)
json_str = json.dumps(data)
print(json_str)
```
